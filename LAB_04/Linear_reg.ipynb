# 1.Importing Libraires

```python
import pandas as pd
```

# 2.Read CSV file

```python
data=pd.read_csv("./data_for_lr.csv")
```

# 3.check the data head

```python
data.head()
```

## 3.1 null values

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

# 4.VIsulization

```python
import matplotlib.pyplot as plt

plt.scatter(data['x'], data['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of data')
plt.show()
```

# 5.Splitting the data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data[['x']], data['y'], test_size=0.2, random_state=42)

print("Training data shape:", X_train.shape, y_train.shape)
print("Testing data shape:", X_test.shape, y_test.shape)

# Import LinearRegression
from sklearn.linear_model import LinearRegression

```

# 6.Model Building and training

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

```

# 7.Model Evaluation and saving

```python
import joblib

accuracy = model.score(X_test, y_test)
print("Model accuracy (R^2 score):", accuracy)
# Save the model to a file
joblib.dump(model, 'linear_regression_model.pkl')
```

# 8.Load the model and predict

```python
loaded_model = joblib.load('linear_regression_model.pkl')
```

## 8.1 Take input from user

```python
X_Pred = int(input("Enter a value to be predicted: "))
```

```python

X_Pred_reshaped = [[X_Pred]]
y_pred = loaded_model.predict(X_Pred_reshaped)


print("Predicted value:", y_pred[0])
```
